# EquiSite Inference

Quick notebook workflow for running per-residue DNA/RNA binding inference from a PDB file.

```python
from pathlib import Path

import torch

from equisite.model import EquiSitePipeline
```

```python
pdb_path = Path("../examples/3HXQ-protein.pdb")
binding_type = "DNA"  # or "RNA"
device = "cuda:0" if torch.cuda.is_available() else "cpu"

pdb_path
```

```python
pipeline = EquiSitePipeline.from_pretrained(
    binding_type=binding_type,
    device=device,
)
pipeline
```

```python
result = pipeline.predict_proba(pdb_path)
top_rows = sorted(result, key=lambda row: row["binding_probability"], reverse=True)[:10]
top_rows
```

```python
output_dir = Path("../examples/notebook_outputs")
output_dir.mkdir(parents=True, exist_ok=True)

result.to_csv(output_dir / f"{pdb_path.stem}_predictions.csv")
result.to_json(output_dir / f"{pdb_path.stem}_predictions.json")

output_dir
```
